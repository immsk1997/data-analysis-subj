```python
import pandas as pd

file = '/content/데이터사이언스_중간고사.csv'
exam = pd.read_csv(file,encoding='utf-8')
exam.head()
```

```python
# 학번 컬럼을 문자열로 형변환
exam['학번'] = exam['학번'].astype(str)
exam.info
```

```python
# 학번 앞 2자리만 추출하여 컬럼값 변경
exam['학번'] = exam['학번'].str.slice(start=0,stop=2)
exam
```

```python
# 컬럼 total > 중간고사_total로 변경
exam1 = exam.rename(columns={'total':'중간고사_total'})
exam1
```

```python
# 신규 컬럼 생성을 위한 정수로의 형변환

exam1.loc[:,'중1'] = exam1.loc[:,'중1'].astype(float)
exam1.loc[:,'중2'] = exam1.loc[:,'중2'].astype(float)
exam1.loc[:,'중3'] = exam1.loc[:,'중3'].astype(float)
exam1.loc[:,'중4'] = exam1.loc[:,'중4'].astype(float)
exam1.loc[:,'중5'] = exam1.loc[:,'중5'].astype(float)
exam1.loc[:,'중6'] = exam1.loc[:,'중6'].astype(float)
exam1.loc[:,'중7'] = exam1.loc[:,'중7'].astype(float)
exam1.loc[:,'중8'] = exam1.loc[:,'중8'].astype(float)
exam1.loc[:,'중9'] = exam1.loc[:,'중9'].astype(float)
exam1.loc[:,'중10'] = exam1.loc[:,'중10'].astype(float)
exam1.loc[:,'중11'] = exam1.loc[:,'중11'].astype(float)
exam1.loc[:,'중12'] = exam1.loc[:,'중12'].astype(float)
exam1.loc[:,'중13'] = exam1.loc[:,'중13'].astype(float)
exam1.loc[:,'중14'] = exam1.loc[:,'중14'].astype(float)
exam1.loc[:,'중15'] = exam1.loc[:,'중15'].astype(float)
exam1.loc[:,'중16'] = exam1.loc[:,'중16'].astype(float)
```

```python
exam1['T/F'] = exam1['중1']+exam1['중2']+exam1['중3']+exam1['중4']+exam1['중5']
```

```python
exam1['빈칸'] = exam1['중6']+exam1['중7']+exam1['중8']+exam1['중9']+exam1['중10']
```

```python
exam1['넘파이'] = exam1['중11']+exam1['중12']+exam1['중13']
```

```python
exam1['판다스'] = exam1['중14']+exam1['중15']
```

```python
exam1['그래프'] = exam1['중16']
```

```python
exam1
```

```python
exam1.loc[:,'산업인공'] = exam1.loc[:,'산업인공'].astype(float)
exam1.loc[:,'프로그랭'] = exam1.loc[:,'프로그랭'].astype(float)
```

```python
# 수강여부 0,1 컬럼 조건문 사용
for Columns in exam1['산업인공']:
  if Columns >1 :
    exam1.loc[exam1['산업인공']>0,'산업인공']='1'
  else:
    exam1.loc[exam1['산업인공']!=exam1['산업인공'],'산업인공']=0

for Columns in exam1['프로그랭']:
  if Columns >1 :
    exam1.loc[exam1['프로그랭']>0,'프로그랭']='1'
  else:
    exam1.loc[exam1['프로그랭']!=exam1['프로그랭'],'프로그랭']=0
```

상관분석 결과 중간고사 총점이 높으면 넘파이, 판다스 문제 총점과의 높은 양의 상관관계가 있다는걸 확인하였습니다.

```python
# 학년 별 과제 총점에 대한 산점도 시각화
import matplotlib.pyplot as plt

x = exam1['학년']
y = exam1['과제1_total']

plt.scatter(x,y,alpha=0.5)
plt.xlabel('Grade')
plt.ylabel('Subject Total')
plt.title("Each Grade's Subject total")
plt.grid()
plt.show()
```

학년별 과제 총점수를 산점도로 시각화 하였으며, 고학년 일수록 과제총점이 높게 분포되어 있는것을 알 수 있었습니다.

```python
# 과제 총점과 중간고사 총점의 히스토그램 시각화
x = exam1['과제1_total']
y = exam1['중간고사_total']
plt.hist(x,bins=10,color='blue')
plt.hist(y,bins=10,color='green')
plt.xlabel('subject total')
plt.ylabel('midterm total')
plt.title("midterm and subject relation")
plt.grid()
plt.show()
```

과제 점수가 높다해서 중간고사 총점이 높은것만은 아니다 라는것을 도출할 수 있었습니다.

```python
# 독립변수 x(넘파이) , 종속변수 y(중간고사 total)와의 선형회귀
from sklearn import linear_model
import numpy as np

x = np.array(exam1['넘파이']).reshape(-1,1)
y = exam1['중간고사_total']

regr = linear_model.LinearRegression()
regr.fit(x,y)

coef = regr.coef_
intercept = regr.intercept_
score = regr.score(x,y)

print('y = {0}*x + {1:2f}'.format(coef, intercept))
print('데이터와 선형 회귀 직선의 관계점수:{0:.1%}'.format(score))
```

```python
# 넘파이와 중간고사 총점의 회귀분석
plt.figure(figsize=(6,4))
plt.scatter(x,y,alpha=0.5)
y_pred = regr.predict(x)
plt.plot(x,y_pred,color='red')

plt.title('regression')
plt.xlabel('numpy')
plt.ylabel('midterm total')
plt.show()
```

회귀분석 결과: 넘파이 점수가 높은 학생일 수록 중간고사 총점이 높게 비례 분포 되어 있는것을 확인할 수 있었습니다. 

```python
# 독립변수 x(판다스) , 종속변수 y(중간고사 total)와의 선형회귀

a = np.array(exam1['판다스']).reshape(-1,1)
b = exam1['중간고사_total']

regr = linear_model.LinearRegression()
regr.fit(a,b)

coef = regr.coef_
intercept = regr.intercept_
score = regr.score(a,b)

print('y = {0}*a + {1:2f}'.format(coef, intercept))
print('데이터와 선형 회귀 직선의 관계점수:{0:.1%}'.format(score))
```

```python
# 판다스와 중간고사 총점의 회귀분석
plt.figure(figsize=(6,4))
plt.scatter(a,b,alpha=0.5)
b_pred = regr.predict(a)
plt.plot(a,b_pred,color='red')

plt.title('regression')
plt.xlabel('pandas')
plt.ylabel('midterm total')
plt.show()
```

회귀분석 결과 : 판다스 점수가 높을수록 중간고사 총점이 높게 비례되어 있는것을 알수 있었습니다.
