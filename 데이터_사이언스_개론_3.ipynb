```python
import pandas as pd
import seaborn as sns 
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = '/content/육군+신체측정정보.csv'
army = pd.read_csv(file,encoding='cp949')
army.head()
```

```python
army.info()
```

```python
army1 = army.rename(columns={'측정 일자':'연도', '가슴 둘레 센티미터':'가슴','소매길이 센티미터':'소매','신장 센티미터':'신장','허리 둘레 센티미터':'허리',
                             '샅높이 센티미터':'샅높이','머리 둘레 센티미터':'머리','발 길이 센티미터':'발','몸무게 킬로그램':'몸무게'})
```

```python
army1
```

```python
# 연도 컬럼(데이터프레임형)을 문자열(str)로 형변환
army1['연도'] = army1['연도'].astype(str)
army1.info
```

```python
#1-1 연도를 네자리 숫자로 변환
army1['연도'] = army1['연도'].str.slice(start=0,stop=4)
army1
```

```python
# 자료 전처리
for Columns in range(2,len(army1.columns)):
    if army1.columns[Columns]=='몸무게':
        word='kg'
    else:
        word='cm'
        
    tmp=[]

    for i in army1.iloc[:,Columns]:
        
        if type(i)==str:
                          
            tmp.append(i.replace(word,'')) 
        else:
            tmp.append(i)
    army1.iloc[:,Columns]= tmp
```

```python
army1
```

```python
tmp=[]

for i in army1['허리']:
    if type(i)==str:
        tmp.append(i.split(' (')[0]) 
    else:
        tmp.append(i)
        
```

```python
army1['허리']=tmp
```

```python
army1
```

```python
# 신장, 몸무게 float형으로 형변환
army1.iloc[:,4] = army1.iloc[:,4].astype(float)
army1.iloc[:,9] = army1.iloc[:,9].astype(float)
army1.iloc[:,2] = army1.iloc[:,2].astype(float)
army1.iloc[:,3] = army1.iloc[:,3].astype(float)
army1.iloc[:,5] = army1.iloc[:,5].astype(float)
army1.iloc[:,6] = army1.iloc[:,6].astype(float)
army1.iloc[:,7] = army1.iloc[:,7].astype(float)
army1.iloc[:,8] = army1.iloc[:,8].astype(float)
```

```python
#1-2 bmi 컬럼 추가
army1['bmi'] = army1['몸무게'] / army1['신장']**2
army1
```

```python
#1-3 bmi판정 컬럼 추가
bmijud = []

for row in army1['bmi']:
  if row <0.00185:
    bmijud.append('underweight')
  elif row >=0.00185 and row <=0.00249:
    bmijud.append('normal')
  elif row >=0.0025 and row<=0.00299:
    bmijud.append('overweight')
  elif row >=0.0030 and row<=0.00349:
    bmijud.append('obese')
  else:
    bmijud.append('extremely obese')
  
army1['bmi판정'] = bmijud
army1
```

```python
# 신체자료 상관분석
army1.corr()

# bmi와 상관관계가 높은 3개인자는 몸무게(0.94) , 가슴(0.8792), 허리(0.8722)로 판정됨
```

```python
# 2-1 연도 별 키의 평균 변화 분석

alz = army1['신장'].groupby(army1['연도']).mean()
alz
```

```python
# 2-2 bmi와 몸무게 산점도

x = army1['몸무게']
y = army1['bmi']

plt.scatter(x,y,alpha=0.5)
plt.title('bmi와 상관관계가 높은 몸무게와의 산점도')
plt.xlabel('몸무게')
plt.ylabel('bmi')
plt.show()
```

```python
#2-2 bmi와 가슴과의 산점도
x = army1['가슴']
y = army1['bmi']

plt.scatter(x,y,alpha=0.5)
plt.title('bmi와 상관관계가 높은 가슴에 관한 산점도')
plt.xlabel('가슴')
plt.ylabel('bmi')
plt.show()
```

```python
#2-2 bmi와 허리와의 산점도

x = army1['허리']
y = army1['bmi']

plt.scatter(x,y,alpha=0.5)
plt.title('bmi와 상관관계가 높은 허리와의 산점도')
plt.xlabel('허리')
plt.ylabel('bmi')
plt.show()
```

```python
#2-3 bmi(종속y), 몸무게(독립x) 선형회귀 분석

from sklearn import linear_model

x = np.array(army1['몸무게']).reshape(-1,1)
y = army1['bmi']

regr = linear_model.LinearRegression()
regr.fit(x,y)

coef = regr.coef_
intercept = regr.intercept_
score = regr.score(x,y)

print('y = {0}*x + {1:2f}'.format(coef, intercept))
print('데이터와 선형 회귀 직선의 관계점수:{0:.1%}'.format(score))
```

```python
#2-4 2-3에 대한 가시화
plt.figure(figsize=(6,4))

x = np.array(army1['몸무게']).reshape(-1,1)
y = army1['bmi']

plt.scatter(x,y,alpha=0.5)

y_pred = regr.predict(x)
plt.plot(x,y_pred,color='red')

plt.title('몸무게와 bmi와의 회귀분석')
plt.xlabel('몸무게')
plt.ylabel('bmi')
plt.show()
```
