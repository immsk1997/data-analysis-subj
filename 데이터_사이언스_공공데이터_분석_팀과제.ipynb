```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 전동퀵보드 연도별 사고현황
data = pd.read_csv('/content/forder/도로교통공단_개인형이동수단(PM) 2017~2019연도별 사건현황 .csv')
data.head()
```

```python
# 결측값 확인
data.isnull().sum()
```

```python
# 결측값 0으로 대체
data = data.fillna(0)
```

```python
data.info()
```

```python
# float로 형변환
data.iloc[:,2] = data.iloc[:,2].astype(float)
data.iloc[:,3] = data.iloc[:,3].astype(float)
data.iloc[:,4] = data.iloc[:,4].astype(float)
data.iloc[:,5] = data.iloc[:,5].astype(float)
data.iloc[:,6] = data.iloc[:,6].astype(float)
data.iloc[:,7] = data.iloc[:,7].astype(float)
data.iloc[:,8] = data.iloc[:,8].astype(float)
data.iloc[:,9] = data.iloc[:,9].astype(float)
data.iloc[:,10] = data.iloc[:,10].astype(float)
```

```python
data.corr()
```

```python
# 서울시 인구밀도 현황
path = pd.read_csv('/content/forder/서울시 각 구별 인구현황.csv')
path.head()
```

```python
path.columns
```

```python
# 각 컬럼별 계산을 위하여 float로 형변환
path.loc[:,'면적'] = path.loc[:,'면적'].astype(float)
path.loc[:,'인구밀도'] = path.loc[:,'인구밀도'].astype(float)
path.loc[:,'인구'] = path.loc[:,'인구'].astype(float)
path.loc[:,'자치구'] = path.loc[:,'자치구'].astype(str)
```

```python
# 자치구 별 인구수 평균
path.groupby(['자치구','인구']).mean()
```

```python
# 자치구 별 인구밀도 평균값
path.groupby(['자치구','인구밀도']).mean()
```

```python
# 결측값 0으로 대체
path = path.fillna(0)
```

```python
path.corr()
```

```python
# 서울시 인구 데이터와 연도별 전동퀵보드 사고현황 csv파일 통합
dtpa = pd.concat([data,path],axis=1)
```

```python
dtpa = dtpa.fillna(0)
```

```python
dtpa
```

```python
# 통합된 데이터 속에서의 서울 자치구 별 2019 사고건수 평균
dtpa.groupby(['자치구','2019 사고건수']).mean()
```

```python
# 서울시 전지역 인구밀도 별 2019 퀵보드 사고현황 평균
dtpa.groupby(['인구밀도','2019 사고건수']).mean()
```

```python
# 서울시 전지역 인구 별 2019 사고건수 평균값
dtpa.groupby(['인구','2019 사고건수']).mean()
```

통합이 어떻게 잘못되었는지 파악이 잘 안됩니다.

```python
dtpa.corr()
```

```python
# 서울시 전체 인구별 2019 전동퀵보드 사고건수에 관한 시각화
x = dtpa['인구']
y = dtpa['2019 사고건수']

plt.scatter(x,y,alpha=0.5)
plt.xlabel('population')
plt.ylabel('2019 accident')
plt.title("accident's population")
plt.grid()
plt.show()
```

서울시의 0~20만 사이의 지역의 사고 건수가 밀집되어 있다는 것을 알수있다

```python
# 독립변수 a(인구) , 종속변수 b(2019 사고건수))와의 선형회귀
from sklearn import linear_model

x = np.array(dtpa['인구']).reshape(-1,1)
y = dtpa['2019 사고건수']

regr = linear_model.LinearRegression()
regr.fit(x,y)

coef = regr.coef_
intercept = regr.intercept_
score = regr.score(x,y)

print('y = {0}*x + {1:2f}'.format(coef, intercept))
print('데이터와 선형 회귀 직선의 관계점수:{0:.1%}'.format(score))
```

```python
# 인구, 2019 사고건수 선형 회귀분석
plt.figure(figsize=(6,4))
plt.scatter(x,y,alpha=0.5)
y_pred = regr.predict(x)
plt.plot(x,y_pred,color='red')

plt.title('regression')
plt.xlabel('population')
plt.ylabel('2019 accident')
plt.show()
```

```python
# 2019 전동퀵보드 사고건수와 서울시 전체 인구밀도와의 산점도 시각화
x = dtpa['인구밀도']
y = dtpa['2019 사고건수']

plt.scatter(x,y,alpha=0.5)
plt.xlabel('population density')
plt.ylabel('2019 accident')
plt.title("accident's population density")
plt.grid()
plt.show()
```

인구밀도가 증가할수록 2019 전동퀵보드 사고건수가 증가하는것을 알수있다.

```python
# 독립변수 x(인구밀도) , 종속변수 y(2019 사고건수))와의 선형회귀
from sklearn import linear_model

x = np.array(dtpa['인구밀도']).reshape(-1,1)
y = dtpa['2019 사고건수']

regr = linear_model.LinearRegression()
regr.fit(x,y)

coef = regr.coef_
intercept = regr.intercept_
score = regr.score(x,y)

print('y = {0}*x + {1:2f}'.format(coef, intercept))
print('데이터와 선형 회귀 직선의 관계점수:{0:.1%}'.format(score))
```

```python
# 인구밀도, 2019 사고건수의 회귀분석
plt.figure(figsize=(6,4))
plt.scatter(x,y,alpha=0.5)
y_pred = regr.predict(x)
plt.plot(x,y_pred,color='red')

plt.title('regression')
plt.xlabel('population density')
plt.ylabel('2019 accident')
plt.show()
```

각 데이터간의 상관분석에는 큰 결함은 없으나,
